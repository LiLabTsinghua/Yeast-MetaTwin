```python
from rdkit import Chem
from rdkit.Chem import rdChemReactions
from rdkit.Chem import Draw

reaction_smarts = "[OH2:3].[O:1]=[CH:2][C@@H:4]([OH:5])[CH2:6][OH:7]>>[O:1]=[C:2]([OH:3])[C@@H:4]([OH:5])[CH2:6][OH:7]"
rxn = rdChemReactions.ReactionFromSmarts(reaction_smarts)
img = Draw.ReactionToImage(rxn, subImgSize=(300, 300))
# img.save('reaction_visualization.png')
img
```
