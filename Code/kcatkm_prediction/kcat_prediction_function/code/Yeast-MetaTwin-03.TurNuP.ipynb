```python
from kcat_prediction import *
from ast import literal_eval
```

```python
# input
yeast8U_sequence_smiles_pre_path = '../../../../Results/kcat_km_predict/yeast8U_sequence_smiles_prepare.csv'

# output
yeast8U_kcat_result_path = '../../../../Results/kcat_km_predict/yeast8U_kcat_result_TurNuP.csv'
```

```python
Kcat_pre_pd = pd.read_csv(yeast8U_sequence_smiles_pre_path)
Kcat_pre_pd = Kcat_pre_pd[['rea_id','gene','reactant_SMILES','product_SMILES','Sequence']]
print(len(Kcat_pre_pd))
Kcat_pre_pd = Kcat_pre_pd.drop_duplicates(subset=['gene','reactant_SMILES','product_SMILES'], keep='first')
print(len(Kcat_pre_pd))
Kcat_pre_pd.head()
```

```python
Kcat_pre_pd['reactant_SMILES'] = Kcat_pre_pd['reactant_SMILES'].apply(lambda x:literal_eval(x))
Kcat_pre_pd['product_SMILES'] = Kcat_pre_pd['product_SMILES'].apply(lambda x:literal_eval(x))
Kcat_pre_pd.head()
```

```python
Kcat_pre_pd['reactant_SMILES'] = Kcat_pre_pd['reactant_SMILES'].apply(lambda x:';'.join(x))
Kcat_pre_pd['product_SMILES'] = Kcat_pre_pd['product_SMILES'].apply(lambda x:';'.join(x))
```

```python
Kcat_pre_pd = Kcat_pre_pd.reset_index(drop=True)
Kcat_pre_pd
```

```python
file_name = f"../result/result_df.csv"
substrates = Kcat_pre_pd['reactant_SMILES'].to_list()
products = Kcat_pre_pd['product_SMILES'].to_list()
enzymes = Kcat_pre_pd['Sequence'].to_list()  
result_df = kcat_predicton(substrates = substrates,
            products = products,
            enzymes = enzymes)   
result_df.to_csv(file_name, index=None)
```
